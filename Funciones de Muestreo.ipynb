#### 2) Muestreo Sistemático:

```python
def systematic_sampling(data, step):
    indexes = np.arange(0, len(data), step=step)
    systematic_sample = data.iloc[indexes] #.iloc[filas, columnas] es para seleccionar filas y/o columnas de un dataframe
    return systematic_sample

systematic_sample = systematic_sampling (data, 3)
systematic_sample
```

Nuestros datos dicen que la proporción es la siguiente:

1) Hoteles en Cuauhtémoc 0.5

2) Museos en Cuauhtémoc 0.2

3) Hoteles en Venustiano Carranza 0.1

4) Mercados en Cuauhtémoc 0.1

5) Mercados en Venustiano Carranza 0.1

Para el muestreo estratificado, vamos a crear una función que nos permita tener las proporciones que hemos dicho anteriormente para cada tipo.
Valores estratificados: 5, los descritos arriba.

```python
def data_estratificado(data, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state=None):
    df_estrat = pd.DataFrame(columns = data.columns) #Crear un dataframe vacío con los nombres de columna del dataset original
    
    pos = -1 # Inicializamos la posición, para que inicie en el 0
    for i in range (len(valores_estrat)): #Iteramos sobre los valores estratificados
        pos +=1
        #Se determina cuántas filas tomar de cada estrato:
        if  pos == len(valores_estrat) -1: #Al llegar a la posición final
            ratio_len = len(data) - len(df_estrat) #Verificamos que el número de filas de entrada = salida
        else:
            ratio_len = int(len(data) * prop_estrat[i])
        #Filtramos el dataset para quedarnos solo con el estrato:
        df_filtrado = data[data[nombres_columnas_estrat] == valores_estrat[i]]
        #Muestreo del estrato:
        df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state = random_state)
        #replace=True: muestreo con reemplazo
        #Se agrega al dataframe acumulado
        df_estrat = pd.concat([df_estrat, df_temp])
    #Devolver el dataframe estratificado    
    return df_estrat 
```

```python
valores_estrat = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel', 'Cuauhtémoc,Mercado', 'Venustiano Carranza,Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1, 0.1]
df_estrat = data_estratificado(data, 'estratificado', valores_estrat, prop_estrat, random_state=42)
df_estrat
```
