### Importamos las librerías necesarias para el análisis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Leemos la base de datos desde un url

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv(url, names=names)
iris.head()
```

```python
#Vamos a graficar la variable sepal-length
plt.hist(iris['sepal-length'], color='orange', bins= int(180/5))
```

Podemos observar que hay menos datos en los extremos de la gráfica y la mayoría se concentra en medio.
Debemos asegurarnos de que se trata de una distribución normal, para ello escribimos el siguiente código:

```python
sns.histplot(iris['sepal-length'], kde=True, bins=int(180/5))
plt.show()
```
