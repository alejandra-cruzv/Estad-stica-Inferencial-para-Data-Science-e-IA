#### Importamos las librerías necesarias para el análisis

```python
import pandas as pd
import numpy as np
import random
```

```python
#Creamos una base de datos.
# Pensemos que se trata de edades, donde el promedio es de 34 años y la muestra es de 10 mil personas
data = np.random.normal(loc=34, size=10000)
data
```

```python
#Hacemos la validación
data.mean()
```

#### Iniciamos Bootstrapping para calcular el promedio de edades

```python
#Vamos a hacer 40 remuestreos, cada uno con 5 elementos
promedio = []
for i in range(40):
    muestra = random.sample(data.tolist(), 5)
    prom = np.mean(muestra)
    promedio.append(prom)
```

```python
#Calculamos el promedio de las 40 muestras
np.mean(promedio)
```

El Bootstrapping puede ser muy efectivo para evitar un sesgo que es muy común en la ciecia de datos y en los modelos de machine learning, conocido como Overfitting. Este concepto nos quiere decir que, el modelo en lugar de aprender de los datos, los está memorizando, y así, en un contexto diferente, el modelo no tiene valor.
