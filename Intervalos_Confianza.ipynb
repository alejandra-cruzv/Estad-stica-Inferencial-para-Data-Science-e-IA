### Importamos las librerías necesarias para el análisis 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import scipy.stats as st
```

### Esta práctica la haremos sobre una distribución creada por nosostros

```python
np.random.seed(20) #Semilla aleatoria de 20
var1 = np.random.randint(0,10,10) #Días de vacaciones
var2 = var1 + np.random.normal(0,1,10) #Dinero gastado
```

```python
print(var1)
print(var2)
```

#### Visualización de resultados con un intervalo de confianza del 80%

```python
grafico = sns.regplot(x=var1, y=var2, ci=80) #Regplot -> Plot de regresión, ci=80 -> Intervalo de confianza del 80%

# Días de vacaciones vs Dinero gastado-> Corelación directa
```

```python
#Unimos las dos variables
tuple = list(zip(var1, var2))
tuple
```

```python
#Creamos una tabla
tabla = pd.DataFrame(tuple,
                     columns=['Días_vacaciones', 'Dinero_gastado'])
tabla
```

#### Calcular intervalos de confianza al 95%

```python
#Utilizamos la distribución t, debido a que se trata solo de una muestra, 
# y no supera los 30 valores; es decir, no podemos hacer uso del T.L.C
st.t.interval(0.95, df=len(tabla)-1, loc=np.mean(tabla), scale=st.sem(tabla))

#scale = error estándar de la media (SEM).
#SEM = desviación estándar / √n.
```

Obtuvimos como intervalos de confianza:

Días de vacaciones: [2.9206, 2.932] -> Es decir, podemos afirmar con un 95% de confianza que las personas salen de vacaciones entre 2 y 3 días.

Dinero gastado: [6.92, 6.93] -> Las personas, gastan aprox. el 69% de su dinero

### **Ejemplo 2**

Se proporcionan los pesos en gramos de 16 cajas de cereal que se seleccionaron de un proceso de llenado con el propósito de verificar el peso promedio.
Determinar un intervalo de confianza al 95%:

```python
x=[506, 508, 499, 503, 504, 510, 497, 512, 514, 505, 493, 496, 506, 502, 509, 496]
```

```python
st.t.interval(0.95, df=len(x)-1, loc=np.mean(x), scale=st.sem(x))
```

I.C.[500.445, 507.054]
