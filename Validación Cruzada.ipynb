Importamos las librerías necesarias

```python
import pandas as pd 
import numpy as np 
import io
```

Importamos el dataset.
El dataset es una base de datos de cáncer, donde nos indica el diagnóstico del paciente (tumor beningo o maligno) y la descripción del tumor encontrado.

```python
url = 'https://raw.githubusercontent.com/alejandra-cruzv/Estad-stica-Inferencial-para-Data-Science-e-IA/refs/heads/master/cancerdata.csv'
data = pd.read_csv(url)
data.head()
```

```python
#Dividir las variables
X = data.iloc[:,2:] #Variables numéricas
Y = data.iloc[:,1]  #Variable categórica
```

```python
X.head()
```

```python
Y.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier #Modelo
from sklearn.model_selection import train_test_split #División de los datos

#Random_state nos ayuda a fijar los conjuntos de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
modelo = DecisionTreeClassifier() #Creación del modelo
modelo.fit(X_train, Y_train)
resultado = modelo.score(X_test, Y_test)
print(resultado) #Imprimir el resultado de accuracy del modelo
```

Iniciamos la validación cruzada, usamos k-fold

```python
from sklearn.model_selection import KFold

modelo = DecisionTreeClassifier() #Definimos el modelo
kfold_validacion = KFold(10) #Creamos los k=10 grupos
```

```python
from sklearn.model_selection import cross_val_score

resultados = cross_val_score(modelo, X, Y, cv=kfold_validacion)
print(resultados)
resultados.mean()
```

Conclusión. Creamos un modelo usando árboles de decision, al aplicar k_folds como método de validación cruzada, ha resultado en valores cercanos al 100%; es decir, el modelo demuestra tener buena precisión.
