### Importamos las librerías necesarias


```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
from IPython.display import Image
```

### Leemos la base de datos

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv(url, names=names)
iris.head()
```

![](petalo-sepalo_0ee5ed06-dbab-456d-a755-a584d0ecb5b4.jpeg)


```python
#Vamos a graficar la variable sepal-length, la cuál vamos a estudiar
plt.hist(iris['sepal-length'], color='orange', bins= int(180/5))
```

Podemos observar que hay menos datos en los extremos de la gráfica y la mayoría se concentra en medio.
Debemos asegurarnos de que se trata de una distribución normal, para ello escribimos el siguiente código:

```python
sns.distplot(iris['sepal-length'], hist=True, kde=True, bins=int(180/5), color='orange')
plt.show()
```

#### Podemos observar que, en efecto, se trata de una población normal.

### Cálculo de la varianza 

![Cálculo de la varianza](formula_1.png)

```python
iris['sepal-length'].var()
```

![](formula2.webp)


```python
iris['sepal-length'].std()
```

Para poder comparar los valores obtenidos de varianza y desviación estándar, vamos a calcular la medida promedio de la variable de interés: 

```python
iris['sepal-length'].mean()
```

Dado que la medida promedio es de aprox. 5.85 y tenemos una varianza de:
Varianza=0.68 aprox. 11% de la medida promedio
Desviación estándar = 0.83 aprox. 14% de la medida promedio

### Creación de una muestra mediante el muestreo aleatorio simple

```python
muestra = iris.sample(frac=0.5)
muestra
```

#### Cálculo de la varianza

#### Cálculo de la desviación estándar
